```python
#wap to find factorial
num=5
factorial=1
if num<0:
    print("factorial not exist")
elif num==0:
    print("factorial of 0 is 1")
else:
    for i in range (1,num+1):
        factorial=factorial*i
print(f"factorial of num {num} is {factorial}")        
```
