```python
num = int(input("Enter a number: "))

# Reverse the number
reverse = 0
temp = num

while temp > 0:
    digit = temp % 10 #get the last digit
    reverse = reverse * 10 + digit
    temp //= 10    #remove the last digit

# Check palindrome
if num == reverse:
    print("Palindrome number")
else:
    print("Not a palindrome")
```

```python
#reverse a number 
num = int(input("Enter a number: "))
reverse = 0
temp = num

while temp > 0:
    digit = temp % 10        # get the last digit
    reverse = reverse * 10 + digit
    temp //= 10              # remove last digit 121//10=12

print("Reversed number:", reverse)
```

```python
num = int(input("Enter a number: "))
temp = num
sum = 0
n = len(str(num))  # number of digits

while temp > 0:
    digit = temp % 10 #get the last number
    sum += digit ** n
    temp //= 10 #remove the last digit

if num == sum:
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
```

```python
#factorial of number
num = int(input("Enter a number: "))
factorial = 1

for i in range(1, num + 1):
    factorial *= i #fact=fact*i

print("Factorial of", num, "is", factorial)
```

```python
n = int(input("Enter the number of terms: "))  #up to 10

a, b = 0, 1  # first two terms
print("Fibonacci Series:", end=" ")

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b  # update a and b for next term
```
